- Use regular expressions to remove problematic zeros from an IPv4 address expressed as a 
string. (example: "216.08.094.196" should become "216.8.94.196", but "216.80.140.196" 
should remain "216.80.140.196"). 

Answer

```python
#https://stackoverflow.com/questions/58013274/python-remove-leading-zeros-from-ip-addresses
def normalize_ip(txt):
    aux = re.sub('(^|\.)0+(?=[^.])', r'\1', txt)
    return aux
#this is a little bit more complex, the section (^|.) indicates that the 0 could be at the begining of the string or after a . the 0+ searches one or
#more zeros, this (?=[^.]) is the lookahead which means that we are looking for zeros that are not at the end of a subdomain and the r'\1' will replace
#the zeros with the beggining of the string or with a .
```

```python
print(normalize_ip("216.0.094.196"))
#print results
```

```python
print(normalize_ip("216.08.094.196"))
#print results
```

```python
print(normalize_ip("216.80.140.196"))
#print results
```

- Use regular expressions to transform a date from MM-DD-YYYY format to DD-MM-YYYY 
format. (example "11-06-2020" should become "06-11-2020"). Optionally*, do the same thing using the `datetime` package.

Answer

```python
#https://www.progress.com/es/blogs/formato-de-fecha-en-python
from datetime import datetime
def switch_md(txt):
    aux = re.sub('(\d{2})-(\d{2})-(\d{4})', r'\2-\1-\3', txt)
    return aux
    
def switch_md_datetime(txt):
    txt = datetime.strptime(txt, '%m-%d-%Y')
    return txt.strftime('%d-%m-%Y')

#in this case the first function we save the three values that have the - between them, the first has to have two digits, the second too and the third
#four, afther that we just re-write them putting the day before the month
#the second functions is the same but with the datetime package, we just take the value first and then return it reorganize it
```

```python
print(switch_md("11-06-2020"))
print(switch_md_datetime("11-06-2020"))
#printing results
```

- Look at the text content and load as xml.

```python
import requests
import xml.etree.ElementTree as ET

proxies = {
    "http": None,
    "https": None,
}
page = requests.get('https://www.w3schools.com/xml/cd_catalog.xml', proxies=proxies)

if page.status_code == 200:
    root = ET.fromstring(page.content)
    stack = [(root, 0)] 
    while stack:
        elem, level = stack.pop()
        print(f"{elem.tag}: {elem.text.strip() if elem.text else 'None'}")
        stack.extend((child, level + 1) for child in reversed(list(elem)))
    
else:
    print(f"Error: {page.status_code}")

#https://docs.python.org/es/3/library/xml.etree.elementtree.html
#in this section we also get the information from the website but instead of printing it right ahead we give some format to it, first we ensure that the
#information was succesfully saved, then we convert the information to an element tree, after that we create a list with a tuple with our element tree
#and its level, then while there are still elements on our tree we take the last element of our list, then with the first label we print the name of the
#label of our actual element  and text (if it is not none we apply the strip method to remove blank spaces, in the other case we print None)
#on the last part we add the childs of the actual element 
```
